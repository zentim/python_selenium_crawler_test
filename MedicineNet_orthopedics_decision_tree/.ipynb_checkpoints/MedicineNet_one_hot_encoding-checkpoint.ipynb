```python
# Load libraries
from sklearn.preprocessing import MultiLabelBinarizer
import numpy as np
```

# Load dataset

```python
import os
import json

# read json files
medicine_net_arr = []

# get dataset full path
dir_path = os.path.dirname(os.path.abspath("MedicineNet_one_hot_encoding.ipynb"))
dataset_path = '/dataset/'
full_path = dir_path + dataset_path
filename = 'MedicineNet_orthopedics_dataset.json'
with open(full_path + filename, 'r') as reader:
    jf = json.loads(reader.read())
    medicine_net_arr += jf

print('Number of diseases : ' + str(len(medicine_net_arr)))
```

```python
for obj in medicine_net_arr:
    print(obj['disease'])
```

```python
disease_name_list = []
obj_symptom_collection = []

for obj in medicine_net_arr:
    # for disease_name_list
    temp = []
    temp.append(obj['disease'])
    disease_name_list.append(temp)
    
    
    # for y
    obj_symptoms = obj['symptom']
    obj_symptom_collection.append(obj_symptoms)
```

```python
print(disease_name_list)
```

```python
print(obj_symptom_collection)
```

```python
# Create MultiLabelBinarizer object
one_hot = MultiLabelBinarizer()

# One-hot encode data
one_hot.fit_transform(obj_symptom_collection)
```

```python
# View classes
one_hot.classes_
```

```python
print(len(one_hot.classes_))
```

```python
import pandas as pd
 
list = one_hot.fit_transform(obj_symptom_collection)
name = one_hot.classes_
test = pd.DataFrame(columns=name,data=list) #  one,two,three
print(test)
test.to_csv(dir_path + '/dataset/MedicineNet_one_hot_encoding.csv',encoding='gbk')
```

```python
list = disease_name_list
name = ['disease']
test = pd.DataFrame(columns=name,data=list) #  one,two,three
print(test)
test.to_csv(dir_path + '/dataset/MedicineNet_one_hot_encoding_disease.csv',encoding='gbk')
```
