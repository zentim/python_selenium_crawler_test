# Load Data

```python
import pandas as pd
```

```python
df_pivoted = pd.read_csv("dataset/MedicineNet_one_hot_encoding_for_decision_tree.csv", encoding ="ISO-8859-1")
```

```python
len(df_pivoted)
```

```python
cols = df_pivoted.columns
```

```python
cols = cols[1:]
```

```python
df_pivoted = df_pivoted.groupby('disease').sum()
df_pivoted = df_pivoted.reset_index()
df_pivoted[:5]
```

```python
len(df_pivoted)
```

```python
x = df_pivoted[cols]
y = df_pivoted['disease']
```

# Trying out our classifier to learn diseases from the symptoms

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.naive_bayes import MultinomialNB
# from sklearn.cross_validation import train_test_split
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
mnb = MultinomialNB()
mnb = mnb.fit(x_train, y_train)
```

```python
mnb.score(x_test, y_test)
```

# Inferences on train and test split

```python
mnb_tot = MultinomialNB()
mnb_tot = mnb_tot.fit(x, y)
```

```python
mnb_tot.score(x, y)
```

```python
disease_pred = mnb_tot.predict(x)
```

```python
disease_real = y.values
```

```python
for i in range(0, len(disease_real)):
    if disease_pred[i]!=disease_real[i]:
        print ('Pred: {0} Actual:{1}'.format(disease_pred[i], disease_real[i]))
```

# Training a decision tree

```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
```

```python
print ("DecisionTree")
dt = DecisionTreeClassifier()
clf_dt=dt.fit(x,y)
print ("Acurracy: ", clf_dt.score(x,y))
```

```python
from sklearn import tree 
from sklearn.tree import export_graphviz

# full tree
export_graphviz(dt, 
                out_file='DOT-files/tree.dot', 
                feature_names=cols,
                class_names=y,
                filled=True, 
                rounded=True,)
```

Go to DOT-files folder to run the cmd: `dot -Tpdf tree.dot -o tree.pdf`

```python
import pydotplus
from IPython.display import Image

# depth 5 tree
dot_data = export_graphviz(dt, 
                            out_file=None, 
                            feature_names=cols,
                            class_names=y,
                            filled=True,
                            rounded=True,
                            #max_depth = 5,
                          )

graph = pydotplus.graph_from_dot_data(dot_data)  
Image(graph.create_png())  
```
