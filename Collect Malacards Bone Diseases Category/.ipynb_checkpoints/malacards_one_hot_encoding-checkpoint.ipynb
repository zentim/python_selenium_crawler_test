```python
# Load libraries
from sklearn.preprocessing import MultiLabelBinarizer
import numpy as np
```

# Load dataset

```python
import os
import json

# read json files
malacards_arr = []

# get dataset full path
dir_path = os.path.dirname(os.path.abspath("malacards_one_hot_encoding.ipynb"))
dataset_path = '/malacards_bone_diseases_category/'
full_path = dir_path + dataset_path

#
for filename in os.listdir(full_path):

    with open(full_path + filename, 'r') as reader:
        jf = json.loads(reader.read())
        malacards_arr += jf

# filter has ICD10's malacard
malacards_has_ICD10_arr = filter(lambda obj: len(obj['ICD10']) != 0, malacards_arr)
for obj in malacards_has_ICD10_arr[0:1]:
    print(obj)

# # split data for each ICD10
# new_malacards_has_ICD10_arr = []
# for obj in malacards_has_ICD10_arr:
#     for obj_ICD10 in obj['ICD10']:
#         temp_obj = obj
#         temp_obj['ICD10'] = obj_ICD10
#         new_malacards_has_ICD10_arr.append(temp_obj['ICD10'])

# print("\nNumber of diseases with ICD10 code: " + str(len(malacards_has_ICD10_arr)))
# print("\nNumber of diseases with ICD10 code (split data for each ICD10): " + str(len(new_malacards_has_ICD10_arr)))
```

```python
print(new_malacards_has_ICD10_arr[11])
```

```python
disease_name_list = []
disease_ICD10_list = []
y = []

for obj in malacards_has_ICD10_arr:
    obj_symptom_collection = obj['sign_or_symptoms'] + obj['HOP_symptoms']
    # remove repeat and turn type to tuple
    obj_symptom_collection = tuple(set(obj_symptom_collection))
    
    # for disease_name_list
    temp = []
    temp.append(obj['name'])
    disease_name_list.append(temp)
    
    # for disease_ICD10_list
    temp = []
    temp += map(lambda x: x[:3], obj['ICD10'])
    disease_ICD10_list.append(temp)
    
    # for y
    y.append(obj_symptom_collection)
```

```python
# Create MultiLabelBinarizer object
one_hot = MultiLabelBinarizer()

# One-hot encode data
one_hot.fit_transform(y)
```

```python
# View classes
one_hot.classes_
```

```python
print(len(one_hot.classes_))
```

```python
import pandas as pd
 
list = one_hot.fit_transform(y)
name = one_hot.classes_
test = pd.DataFrame(columns=name,data=list) #  one,two,three
print(test)
test.to_csv(dir_path + '/malacards_one_hot_encoding.csv',encoding='gbk')
```
